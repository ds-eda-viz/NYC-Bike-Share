```python
import pandas as pd
```

```python
all_location = pd.read_csv('all_station_location.csv', dtype=str)
```

```python
all_results = []
```

```python
all_location.iloc[12][1]
```

```python
for a, b in zip(all_location['latitude'], all_location['longitude']):
    print(a, b)
    response = requests.get('https://maps.googleapis.com/maps/api/geocode/json?latlng={},{}&key={}&result_type=administrative_area_level_2|sublocality_level_1'
                        .format(a, b, api_key))

    resp_json_payload = response.json()
    all_results.append(resp_json_payload)

    print(resp_json_payload['results'][0]['address_components'][0]['short_name'])
    print(resp_json_payload['results'][0]['address_components'][1]['short_name'])
```

```python
len(all_results)
```

```python
boroughs = []
counties = []
```

```python
for resp_json_payload in all_results:
    boroughs.append(resp_json_payload['results'][0]['address_components'][0]['short_name'])
    counties.append(resp_json_payload['results'][0]['address_components'][1]['short_name'])
```

```python
all_location['borough'] = boroughs
```

```python
all_location['countie'] = counties
```

```python
all_location
```

```python
all_location.to_csv('all_station_location_with_county.csv', index=0)
```
